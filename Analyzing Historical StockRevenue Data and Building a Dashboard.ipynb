```python
import yfinance as yf

# Extract Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_stock_data = tesla.history(period="max")

# Display the data
print(tesla_stock_data.head())
```

```python
import requests
from bs4 import BeautifulSoup

# URL for Tesla revenue data (example URL, replace with actual)
url = "https://example.com/tesla-revenue"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Parse the data (customize based on actual webpage structure)
revenue_data = soup.find_all('some_tag', class_='some_class')

# Display the data
for data in revenue_data:
    print(data.text)
```

```python
# Extract GameStop stock data
gamestop = yf.Ticker("GME")
gamestop_stock_data = gamestop.history(period="max")

# Display the data
print(gamestop_stock_data.head())
```
